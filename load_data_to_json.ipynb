```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import json
```

```python
countries_df=pd.read_csv('Resources/country_comparison_large_dataset.csv')
countries_df
```

```python
row1 = {}
filtered_df = countries_df.drop(columns=['Country', 'Year'])
stats = filtered_df.columns
for stat in stats:
    earliest = countries_df.loc[(countries_df['Country'] == 'USA') & (countries_df['Year'] == 2000), stat]
    latest = countries_df.loc[(countries_df['Country'] == 'USA') & (countries_df['Year'] == 2023), stat]
    row1[stat] = ((latest.iloc[0] - earliest.iloc[0])/earliest.iloc[0])*100
row1['Country'] = 'USA'
new_row_df = pd.DataFrame([row1])
new_row_df
```
